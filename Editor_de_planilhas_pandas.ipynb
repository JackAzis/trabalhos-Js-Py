```python
import pandas as pd
from IPython.display import clear_output


teste = {'nome':['Artur','Alex','Vanessa','Fernanda','Ingrid','Ana','Paulo','David','Mauricio','Alexandre'],'idade':[30,32,40,25,50,20,68,78,29,37],'salario':[8000,5800,2800,9800,6700,4789,5980,6975,9850,7000],'estado':['RJ','SP','RJ','SP','RJ','SP','RJ','SP','RJ','SP'],}
df = pd.DataFrame(teste)
df


def selecao():

  print('Bem vindo ao editor de planilhas\nBy:Paulo David\n\n ')
  selecionado=int(input("""Digite 1 para adicionar colunas
Digite 2 para apagar colunas
Digite 3 para selecionar colunas
Digite 4 para ver as colunas com espaços nulos\n"""))
  match selecionado:
    case 1:
      adicionar_colunas()
    case 2:
      apagar_colunas()
    case 3:
      selecionar_colunas()
    case 4:
      colunas_com_espacos_nulos()
    case _:
      print('Opção inválida')
      selecao()

def adicionar_colunas():
  colunas_da_tabela()
  print('Menu de adição de colunas')
  nova_coluna=input('Digite o nome da nova coluna: ')
  print(f'A coluna, {nova_coluna} foi adicionada')
  df[nova_coluna] = None
  colunas_da_tabela()
  retornar_selecao()

def selecionar_colunas():
    colunas_da_tabela()
    print("\nMenu de seleção de colunas")
    print("1 - Selecionar colunas específicas")
    print("2 - Exibir apenas colunas numéricas")
    print("3 - Exibir apenas colunas de texto")
    print("4 - Filtrar colunas por um valor específico")
    print("5 - Ordenar colunas por valores")
    print("6 - Exportar colunas selecionadas para CSV")

    opcao = input("\nDigite a opção desejada: ")

    if opcao == "1":
        colunas = input('Digite os nomes das colunas separadas por vírgula: ').split(',')
        colunas = [col.strip() for col in colunas]
        try:
            print(df[colunas])
        except KeyError as e:
            print(f'Erro: Algumas colunas não foram encontradas. {e}')

    elif opcao == "2":
        colunas_numericas = df.select_dtypes(include=['number'])
        print(colunas_numericas)

    elif opcao == "3":
        colunas_texto = df.select_dtypes(include=['object'])
        print(colunas_texto)

    elif opcao == "4":
        coluna_filtro = input("Digite o nome da coluna para filtrar: ")
        if coluna_filtro in df.columns:
            valor = input(f"Digite o valor a ser buscado na coluna {coluna_filtro}: ")
            filtro = df[df[coluna_filtro].astype(str) == valor]
            print(filtro)
        else:
            print("Coluna não encontrada.")

    elif opcao == "5":
        coluna_ordenar = input("Digite o nome da coluna para ordenar: ")
        if coluna_ordenar in df.columns:
            ordem = input("Digite 'asc' para crescente ou 'desc' para decrescente: ").lower()
            ascending = True if ordem == 'asc' else False
            df_ordenado = df.sort_values(by=coluna_ordenar, ascending=ascending)
            print(df_ordenado)
        else:
            print("Coluna não encontrada.")

    elif opcao == "6":
        colunas = input('Digite os nomes das colunas separadas por vírgula: ').split(',')
        colunas = [col.strip() for col in colunas]
        try:
            df[colunas].to_csv("dados_selecionados.csv", index=False)
            print("Arquivo 'dados_selecionados.csv' foi salvo com sucesso!")
        except KeyError as e:
            print(f'Erro: Algumas colunas não foram encontradas. {e}')

    else:
        print("Opção inválida!")

    retornar_selecao()

def colunas_com_espacos_nulos():
  try:
    colunas_com_nulos = df.columns[df.isnull().any()]
    print(df[colunas_com_nulos])
  except:
    print('Não há colunas com espaços nulos')
  retornar_selecao()

def apagar_colunas():
  colunas_da_tabela()
  try:
    apagar=input('Digite o nome da coluna que deseja apagar: ')
    del df[apagar]
  except:
    print(f'A coluna, {apagar} não foi encontrada, verifique o nome e tente novamente')
    apagar_colunas()
  retornar_selecao()

def colunas_da_tabela():
  clear_output()
  print(f'As colunas desta tabela são', df.columns.tolist())

def retornar_selecao():
  input('\nDigite qualquer botão para retornando ao menu de seleção: ')
  clear_output()
  selecao()

selecao()


```
